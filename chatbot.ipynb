```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ['OPENAI_API_KEY'] = os.getenv('OPENAI_API_KEY')
os.environ['LANGCHAIN_API_KEY'] = os.getenv('LANGCHAIN_API_KEY')

os.environ['LANGCHAIN_TRACING_v2'] = "true"
os.environ['LANGCHAIN_PROJECT'] = os.getenv("LANGCHAIN_PROJECT")
```

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-4o-mini")
print("Working!")
```

```python
ans = llm.invoke("Who are you and who are better than you?")
print(ans)
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "You are an AI Engineer, provide ans based on questions"),
        ("user", "{input}")
    ]
)
```

```python
chain = prompt | llm

response = chain.invoke({"input", "Who will win today? RCB or CSK"})
print(response)
```

```python
from langchain_core.output_parsers import StrOutputParser
parser = StrOutputParser()
chain = prompt | llm | parser
```

```python
response = chain.invoke({"input", "explain output parser in langchain-core"})
print(response)
```
